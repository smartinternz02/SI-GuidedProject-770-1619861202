```python
def t_marital(a):
    if a=="married":
        return 0
    if a=="single":
        return 1
    if a=="divorced":
        return 2
```

```python
dataset["marital"]=dataset["marital"].apply(t_marital)
dataset
```

```python
def t_job(a):
    if a=="admin.":
        return 0
    if a=="technician":
        return 1
    if a=="services":
        return 2
    if a=="management":
        return 3
    if a=="retired":
        return 4
    if a=="blue-collar":
        return 5
    if a=="unemployed":
        return 6
    if a=="entrepreneur":
        return 7
    if a=="housemaid":
        return 8
    if a=="unknown":
        return 9
    if a=="self-employed":
        return 10
    if a=="student":
        return 11
```

```python
dataset["job"]=dataset["job"].apply(t_job)
dataset
```

```python
def t_edu(a):
    if a=="secondary":
        return 0
    if a=="tertiary":
        return 1
    if a=="primary":
        return 2
    if a=="unknown":
        return 3
```

```python
dataset["education"]=dataset["education"].apply(t_edu)
dataset
```

```python
def t_con(a):
    if a=="unknown":
        return 0
    if a=="cellular":
        return 1
    if a=="telephone":
        return 2
```

```python
dataset["contact"]=dataset["contact"].apply(t_con)
dataset
```

```python
def t_mon(a):
    if a=="jan":
        return 0
    if a=="feb":
        return 1
    if a=="mar":
        return 2
    if a=="apr":
        return 3
    if a=="may":
        return 4
    if a=="jun":
        return 5
    if a=="jul":
        return 6
    if a=="aug":
        return 7
    if a=="sep":
        return 8
    if a=="oct":
        return 9
    if a=="nov":
        return 10
    if a=="dec":
        return 11
```

```python
dataset["month"]=dataset["month"].apply(t_mon)
dataset
```

```python
def t_out(a):
    if a=="unknown":
        return 0
    if a=="other":
        return 1
    if a=="failure":
        return 2
    if a=="success":
        return 3
```

```python
dataset["poutcome"]=dataset["poutcome"].apply(t_out)
dataset
```

```python
dataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dataset["default"]=le.fit_transform(dataset["default"])
dataset["deposit"]=le.fit_transform(dataset["deposit"])
dataset["housing"]=le.fit_transform(dataset["housing"])
dataset["loan"]=le.fit_transform(dataset["loan"])
dataset
```

```python
x=dataset.iloc[:,0:16].values
y=dataset.iloc[:,16].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
ypred=lr.predict(x_test)
ypred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(ypred,y_test)
accuracy
```
